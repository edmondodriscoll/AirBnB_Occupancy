```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from datetime import datetime, date, time


import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
Project_B_Dummied4=pd.read_csv('project_B2_dummied.csv')
```

# Seeing the distrubution of occupancy Rates

```python
# Histogram of occpancy rates ,negatively skewed. 
sns.distplot( Project_B_Dummied4["acc_ave_month_occ"], bins=20 )
```

# Seeing what correlates the most  

```python
# Creating a mask to easily what has the best correlations.
maskCorr=(Project_B_Dummied4.corr()>.1)|(Project_B_Dummied4.corr()<-.1)
maskeddummy=Project_B_Dummied4.corr()[maskCorr]
maskeddummy[['acc_ave_month_occ']].plot(kind='barh',figsize=(120,120),fontsize=60)
plt.show()
```

# Scatter plots 

```python
# Seeing how the average price with acc_ave_month_occ looks on a scatter plot
sns.jointplot(Project_B_Dummied4.acc_ave_month_occ, 
             Project_B_Dummied4.av_price, 
            joint_kws=dict(s=25, alpha=0.6));
```

```python
sns.jointplot(Project_B_Dummied4.acc_ave_month_occ , 
             Project_B_Dummied4.number_of_reviews, 
            joint_kws=dict(s=25, alpha=0.6));
```
