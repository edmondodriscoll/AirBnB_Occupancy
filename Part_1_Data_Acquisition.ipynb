# Step 4, Restructering the scraped calander data

### Once the data has been scraped then it is restructed so it can be more easily grouped 

```python
from datetime import datetime, date, time
now = datetime.now()
today = now.strftime("%Y-%m-%d")
# Organizing the scrapped data into a list 
list_oflist=[]
from tqdm import tqdm

for i in tqdm(range(len(master))):
    try:
        dfss = pd.DataFrame(master[i]['calendar_months'][0]['days'])
        list2 = [x for x in range(len(dfss))]
        index_list= pd.DataFrame(iD_list[i],index=list2, columns=['ID'])
        datetoday= pd.DataFrame(now.strftime("%Y-%m-%d"),index=list2, columns=['date_scraped'])
        result = dfss.join(index_list)
        print(result)
        typee =[]
        local_adjusted_price=[]
        for day in range(len(dfss)):
            
            try:
                fret=master[i]['calendar_months'][0]['days'][day]['price']
                typee.append(fret['type'])
                local_adjusted_price.append(fret['local_adjusted_price'])
                
            except:
                typee.append('Na')
                local_adjusted_price.append('Na')
        
        
        new_df = pd.DataFrame(columns = ['local_adjusted_price','type'])
        new_df['local_adjusted_price']=local_adjusted_price
        new_df['type']=typee
        result2= result.join(new_df)
        result3= result2.join(datetoday)
        list_oflist.append(result3)

    except:
        pass
```

```python
# This groups the list_oflist into a super list
project_b1=pd.concat(list_oflist)
```
